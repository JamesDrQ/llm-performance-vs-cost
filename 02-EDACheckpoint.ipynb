## Research Question

**Can benchmark performance across different domains (reasoning, coding, mathematics, and general knowledge) reliably predict the token pricing of AI language models?**

Subquestions might include: 

- Which models deliver the best overall performance across benchmarks?

- Which models provide the most cost-efficient performance?

## Background and Prior Work

Large language models (LLMs) are largely evaluated using standardized benchmarks to measure performance in reasoning, coding, mathematics, and general knowledge. For example, some widely used tasks include MMLU for academic knowledge, GSM8K for multi-step math reasoning, and HumanEval for code generation. These benchmarks give each model an overall accuracy score that can be compared across multiple systems. In addition to these static tests, preference-based evaluations such as the LMSYS Chatbot Arena use large-scale human comparisons to capture perceived quality across diverse prompts, offering a much broader and holistic view of model performance.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Alongside performance data, providers also publish detailed token pricing, usually expressed as cost per million input and output tokens. OpenAI, Google, and other companies differentiate their models by capability tier, context window, and modality, which results in a pricing structure that is viewable to the general public, however interpreting and analyzing this data is much harder.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) While pricing is public, it is not obvious whether higher benchmark scores actually justify higher token costs. Some analyses suggest that cost-efficiency varies widely across models, and that smaller or open-source systems can sometimes deliver better “performance per dollar” on specific tasks compared to their larger counterparts.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

The findings of this project will be able to help both students and researchers. Students in college span a wide variety of majors, and in todays age, they now mostly all use LLMs in their lives. However, each students’ use for LLMs differs and hence finding the best model for their task is critical, for example: computer science students may care most about coding accuracy, math and statistics students about problem-solving, and social science or humanities students about writing quality and general reasoning. Furthermore, many students also work within a limited budget. If one model is cheaper but nearly as accurate as a flagship model for a given domain, it may be the better practical choice. Moreover, understanding whether or not newer models are actually becoming more cost-efficient and which models are best suited to particular types of tasks could help students pick tools that fit both their needs and their budgets.

*Past work examples:*

Past work has begun to explore model comparisons in more applied settings. For example, articles and blog posts can be found directly comparing “ChatGPT vs DeepSeek,” noting that one system may feel better for general use while another excels on more technical tasks.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) There are even ongoing experiments that let different AI systems trade a virtual portfolio to compare their performance in financial markets over time.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) These examples highlight that “best” depends heavily on the task at hand.


Prior work on LLM evaluation has shown both the strengths and limits of current approaches. For example, the Hugging Face Open LLM Leaderboard combines results from many benchmarks into a single view, which helps make trade-offs between accuracy, efficiency, and model size easier to see.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Other reviews of evaluation methods argue that models should be judged along several dimensions, since no single benchmark can capture everything an LLM can do.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) 

Building on this, our project will compare large language models ranked on public leaderboards across multiple evaluation domains with their published pricing structures. We will then fit a predictive model using benchmark scores as independent variables and token cost as the dependent variable, in order to test whether performance can reliably explain or predict pricing.

## Hypothesis


**Null Hypothesis:** There is no significant relationship between a LLM model’s benchmark accuracy score and its token cost. Improvements in performance do not systematically result in higher or lower token cost.

**Alternative Hypothesis:** A LLM model's benchmark accuracy is significantly related to token cost — either positively (higher performance requires more cost) or non-linearly (diminishing returns).

## **Dataset Column Descriptions:**
---
### **Basic Model Information**
- **id** – Unique numerical identifier for each model.  
- **name** – Human-readable name of the model (e.g., “GPT-5-mini”).  
- **slug** – URL-friendly identifier for the model.  
- **release_date** – Date the model was released.

### **Speed & Latency Metrics**
- **median_output_tokens_per_second** – Model generation speed in tokens per second (higher = faster).  
- **median_time_to_first_token_seconds** – Time until the first generated token appears (seconds).  
- **median_time_to_first_answer_token** – Time until the model begins producing its final answer.

### **Creator Metadata**
- **model_creator.id** – Identifier for the organization that developed the model.  
- **model_creator.name** – Name of the model’s creator (e.g., OpenAI, Anthropic).  
- **model_creator.slug** – URL-friendly version of the creator name.

---
### **Performance Metrics (Evaluation Scores)**  
All scores represent benchmark accuracy or composite abilities; higher values indicate better performance.

### **Composite Scores**
- **evaluations.artificial_analysis_intelligence_index** – Overall composite reasoning score.  
- **evaluations.artificial_analysis_coding_index** – Composite coding/programming ability score.  
- **evaluations.artificial_analysis_math_index** – Composite mathematical reasoning score.

### **Academic / Professional Benchmarks**
- **evaluations.mmlu_pro** – Accuracy on the professional-level MMLU-Pro exam.  
- **evaluations.gpqa** – Accuracy on the graduate-level GPQA benchmark.  
- **evaluations.hle** – High-Level Evaluation score for advanced reasoning.  
- **evaluations.livecodebench** – Coding performance on real-time coding tasks.  
- **evaluations.scicode** – Scientific-reasoning benchmark accuracy.  
- **evaluations.math_500** – Accuracy on 500 diverse math problems.  
- **evaluations.aime** – Score on AIME (American Invitational Math Exam).  
- **evaluations.aime_25** – Score on the 2025 AIME-style problem set.  
- **evaluations.ifbench** – Logical inference (IFBench) reasoning accuracy.  
- **evaluations.lcr** – LCR reasoning benchmark score.  
- **evaluations.terminalbench_hard** – Performance on hard terminal-based coding tasks.  
- **evaluations.tau2** – Score on the TAU2 advanced reasoning benchmark.
---

### **Pricing Metrics**  
Prices are in **USD per 1,000,000 tokens**.

- **pricing.price_1m_blended_3_to_1** – Estimated cost assuming a 3:1 input-to-output token ratio.  
- **pricing.price_1m_input_tokens** – Cost per one million input (prompt) tokens.  
- **pricing.price_1m_output_tokens** – Cost per one million output (generated) tokens.
---
### **Major Concerns**
A major concern with this dataset is that many benchmark scores are missing for a large number of models. Several evaluation suites—such as AIME, IFBench, LCR, TAU2, and TerminalBench—only test a small subset of high-end or well-known models, meaning mid-tier models appear to have missing values not because they perform poorly, but simply because they were never evaluated. This selective inclusion introduces bias and makes direct comparison across all models difficult. Another issue is that different benchmarks are run by different organizations using slightly different prompting setups, evaluation conditions, or decoding parameters, which reduces cross-benchmark consistency. Additionally, token pricing is influenced by company strategy and business decisions rather than purely by model performance, so price should not be interpreted as an objective reflection of capability. Finally, the dataset only includes models that are publicly listed and evaluated through certain platforms, which excludes many open-source or privately deployed models and may skew the sample toward commercial systems from major AI labs.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reload the cleaned dataset:
leaderboard = pd.read_csv('data/02-processed/leaderboard_clean.csv')
leaderboard = leaderboard.set_index('name')
leaderboard.columns = ['release_date', 'creator','intelligence_index', 
                      'hle', 'mmlu_pro', 'gpqa', 'livecodebench', 'scicode', 
                      'blended_price', 'input_price', 'output_price']
leaderboard
```

Our goal with this EDA is to get a sense of the shape of the data, the range of model performances, and whether or not, there are any obvious patterns between benchmark scores and token pricing, and hence draw and predict a possible correlation.

We hence decided to utilise a few different of visualisations to best present the data:

### 1. Histograms

To better understand the distributional shape of our variables, we plotted histograms for key performance and pricing features. A histogram allows us to see whether the data are symmetric, skewed, multimodal, or contain outliers—patterns that summary statistics alone cannot capture.

From the histograms, we can observe the approximate range of values, how densely observations cluster in different regions, and whether the variable follows an approximately normal distribution or exhibits long tails. Understanding these distributional shapes informs later modeling decisions: skewed variables may require transformations, heavy-tailed variables may need robust estimators, and multimodal distributions might suggest meaningful sub-groups among models.

#### Token Pricing
The token prices are the very first variables we pay attention to. Here we first generate histograms to observe their distributions.

```python
prices = ['blended_price', 'input_price', 'output_price']
colors = ['orange','red','yellow']
for i in range(len(prices)):
    plt.figure()
    plt.hist(leaderboard.get(prices[i]), bins=15, color=colors[i], edgecolor='white')
    plt.title(f"Distribution of {prices[i]}", fontsize=15)
    plt.xlabel(prices[i], fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
```

Each Distribution histogram be that blended price, input price, and output price, show us a very consistent and repeating pattern. As we can see, all three distributions are heavily right skewed, with the majority of models clustered at the low end of the cost range, and only a small number in the much higher price levels. 

The blended price and input price histograms both show that most models tend to near the zero end of the spectrum, suggesting that affordable models take up the vast majority of the leaderboard. The output prices however, display the widest spread, with a few models as we can see, charging a large amount more for generation than the rest. Overall, these histograms show that while expensive models do exist, they are clear outliers in this dataset, with most models remaining relatively low cost across all the pricing dimensions.

To improve model accuracy, we first decide to remove the outliers using the IQR Rule. Due to the apparent rightward skewness of the distributions, we only need to deal with upper outliers. That is, any value larger than Q3 + 1.5 IQR should be eliminated as outliers. Moreover, since blended price is the most comprehensive price metric, it's the only variable we are going to deal with.

```python
q1_bp = leaderboard['blended_price'].quantile(0.25)
q3_bp = leaderboard['blended_price'].quantile(0.75)
iqr_bp = q3_bp - q1_bp
upper_bp = q3_bp + 1.5 * iqr_bp

print(upper_bp)
```

```python
# extreme values:
leaderboard[leaderboard['blended_price'] > upper_bp ].get(prices)
```

However, we notice that we are removing some of the most important, state-of-the-art, enterprise-grade LLMs on the market. This means that although the IQR rule is mathematically correct, it's not appropriate for our dataset.

We decide to only remove the outliers that are far larger than the rest in the dataset, since these extreme values disproportionately skew the distribution and obscure meaningful comparisons among typical models. By retaining moderately high values while excluding only the most extreme cases, we preserve the integrity of the dataset while still reducing distortion.

```python
# extreme outliers:
leaderboard[leaderboard['blended_price'] > 25]
```

```python
# remove extreme outliers:
leaderboard = leaderboard[leaderboard['blended_price'] < 25]
```

```python
# regenerate histograms for price:
for i in range(len(prices)):
    plt.figure()
    plt.hist(leaderboard.get(prices[i]), bins=15, color=colors[i], edgecolor='white')
    plt.title(f"Distribution of {prices[i]}", fontsize=15)
    plt.xlabel(prices[i], fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
```

After removing the extreme outliers, we can see that all three histograms are still skewed to the right, but we can see the distribution better now. The spread of blended_price is reduced from (0, 30) to (0, 7). Their centers are likely to be located between (0, 1). Note that the bar at price=0 is very high, suggesting that a lot of tokens cost between (0, 1). 

```python
benchmarks = ['hle', 'mmlu_pro', 'gpqa', 'livecodebench', 'scicode']
colors = ['green','cyan','blue','purple','brown']
for i in range(len(benchmarks)):
    data = leaderboard.get(benchmarks[i])
    mean = np.mean(data)
    median = np.median(data)
    plt.figure()
    plt.title(f"Distribution of {benchmarks[i]}", fontsize=15)
    plt.xlabel(benchmarks[i], fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.hist(data, bins=15, range=(0,1), color=colors[i], edgecolor='white');
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
    plt.axvline(median, color='black', linestyle='--', label=f'Median = {median:.2f}')
    plt.legend()
```

#### hle
- As we can see most models score extremely low on this benchmark, with the distribution being largely packed near the bottom, the median (0.05) would be a better measurement than the mean (0.07), showing that almost all models were struggling here.
#### mmlu_pro
- Scores here are much higher and much more consistent. As we can see, Most models lie between ranges of 0.6 and 0.8, and the mean of 0.66 and the median of 0.73 shows strong overall performance across the board on this benchmark.
#### gpqa
- This distribution is wider, with scores spread from low to high. The mean around 0.54 shows moderate performance, with a lot of variation among models.
#### livecodebench
- Scores vary a lot, ranging from very low to quite high. The mean of ~0.39 suggests this coding benchmark is challenging for many models.
#### scicode
- Most scores fall on the lower end, with the mean around 0.25. This indicates that scientific reasoning tasks are difficult for current models.

From obervation, we determine that there's no need to remove outliers for the performance metric distributions, because they look consistent, and there's no obvious outliers.

```python
plt.hist(leaderboard['intelligence_index'], bins=15, edgecolor='white');
plt.title('intelligence index', fontsize=15);
mean = np.mean(leaderboard['intelligence_index'])
median = np.median(leaderboard['intelligence_index'])
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.axvline(median, color='black', linestyle='--', label=f'Median = {median:.2f}')
plt.legend();
```

#### intelligence index:
Scores cluster around the middle, with a mean of ~30. This index is more balanced, showing moderate performance overall with a smoother spread.

#### 2. Violin plots
These were used to compare how each model performed across different benchmarks (MMLU Pro, GPQA, SciCode, etc.). These were most ideal because they show not only the middle average of the distribution but also the full density of the scores. Allowing us to quickly see which benchmarks are harder, which ones have more spread, and whether many models are hitting upper or lower limits on any particular task.

```python
data = pd.melt(leaderboard[benchmarks])
data.columns = ['benchmark', 'accuracy']
plt.figure(figsize=[10,5])
sns.violinplot(data, hue='benchmark', y='accuracy', palette=colors);
```

The violin plot gives us a quick side by side view of how model accuracy varies across the five different benchmarks. HLE has the lowest and tightest distribution, telling us that almost all models scored poorly there. MMLU-pro stands out with high, consistent scores, and a very little amount of spread. GPQA also performs relatively well, but with more variation across the models. LiveCodeBench has a much wider spread, signalling that each model differs a lot in coding performance. Lastly, SciCode sits in the lower-middle range, showing modest scores with moderate variability.

#### mmlu_pro vs blended_price

#### gpqa vs blended_price

#### livecodebench vs blended_price

#### scicode vs blended_price

#### 4. Pair plots
Last but not least we used pair plots, because the leaderboard includes a large variety of different benchmarks, we generated a pair plot to compare all of them easily at once. This helps us immediately see if any benchmarks are strongly correlated with each other, or whether they capture different aspects of model capability, giving us a broad way to compare each benchmark and to spot any immediate correlations or relationships.

```python
# pairplot
sns.pairplot(leaderboard, vars=benchmarks);
```

The pairplot shows that all of the benchmark scores are strongly positively correlated with each other, with models that perform well on one test tending to perform well on the others too.

Although there were some differences in scale, the scatterplots all generally show the same upward trend, hence we can conclude that these benchmarks generally capture a similar underlying model ability.

A few clusters and curved patterns appear for some pairs, which suggests that the relationships aren’t perfectly linear, but the overall pattern still clearly points to higher scores going together across benchmarks.
